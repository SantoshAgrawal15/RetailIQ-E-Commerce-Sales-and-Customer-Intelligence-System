```python
import pandas as pd
import sqlite3

# Load CSVs into DataFrames
customers = pd.read_csv("customers.csv")
products = pd.read_csv("products.csv")
orders = pd.read_csv("orders.csv")
order_items = pd.read_csv("order_items.csv")
delivery = pd.read_csv("delivery.csv")

# Create in-memory SQLite DB
conn = sqlite3.connect(":memory:")

# Load into tables
customers.to_sql("Customers", conn, index=False, if_exists="replace")
products.to_sql("Products", conn, index=False, if_exists="replace")
orders.to_sql("Orders", conn, index=False, if_exists="replace")
order_items.to_sql("Order_Items", conn, index=False, if_exists="replace")
delivery.to_sql("Delivery", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT
  c.Customer_ID,
  SUM(oi.Quantity * p.Price * (1 - oi.Discount)) AS Total_Spend,
  CASE
    WHEN SUM(oi.Quantity * p.Price * (1 - oi.Discount)) >= 10000 THEN 'High'
    WHEN SUM(oi.Quantity * p.Price * (1 - oi.Discount)) >= 5000 THEN 'Mid'
    ELSE 'Low'
  END AS Segment
FROM Customers c
JOIN Orders o ON c.Customer_ID = o.Customer_ID
JOIN Order_Items oi ON o.Order_ID = oi.Order_ID
JOIN Products p ON oi.Product_ID = p.Product_ID
WHERE o.Status = 'Completed'
GROUP BY c.Customer_ID
ORDER BY Total_Spend DESC;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
  p.Product_Name,
  p.Category,
  SUM(oi.Quantity * p.Price * (1 - oi.Discount)) AS Total_Revenue
FROM Order_Items oi
JOIN Products p ON oi.Product_ID = p.Product_ID
JOIN Orders o ON o.Order_ID = oi.Order_ID
WHERE o.Status = 'Completed'
GROUP BY p.Product_Name, p.Category
ORDER BY Total_Revenue DESC
LIMIT 10;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
  o.Order_ID,
  o.Customer_ID,
  o.Status,
  d.Delivery_Status
FROM Orders o
LEFT JOIN Delivery d ON o.Order_ID = d.Order_ID
WHERE o.Status != 'Completed' OR d.Delivery_Status != 'Delivered';
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
  o.Order_ID,
  o.Order_Date,
  d.Delivery_Date,
  JULIANDAY(d.Delivery_Date) - JULIANDAY(o.Order_Date) AS Delivery_Days,
  CASE
    WHEN JULIANDAY(d.Delivery_Date) - JULIANDAY(o.Order_Date) <= 7 THEN 'On-Time'
    ELSE 'Delayed'
  END AS Delivery_Status
FROM Orders o
JOIN Delivery d ON o.Order_ID = d.Order_ID
WHERE o.Status = 'Completed';
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
  STRFTIME('%Y-%m', o.Order_Date) AS Month,
  ROUND(SUM(oi.Quantity * p.Price * (1 - oi.Discount)), 2) AS Monthly_Revenue
FROM Orders o
JOIN Order_Items oi ON o.Order_ID = oi.Order_ID
JOIN Products p ON oi.Product_ID = p.Product_ID
WHERE o.Status = 'Completed'
GROUP BY Month
ORDER BY Month;
"""
pd.read_sql_query(query, conn)
```

```python
result = pd.read_sql_query(query, conn)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(result['Month'], result['Monthly_Revenue'], marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
